```python
from herramientas.colorizer.continuous_colorizer import ContinuousColorizer
from herramientas.colorizer.paleta_argendata import *
from herramientas.colorizer.palette import Palette
from herramientas.colorizer.paleta_fundar import *
from herramientas.colorizer.basic_colors import *
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

import numpy as np
import polars as pl

colores_por_presidencia = dict(
    fdelarua_0      =   '#8c564b', # tab:brown
    eduhalde_0      =   '#d62728', # tab:red
    nkirchner_0     =   '#1f77b4', # tab:blue
    ckirchner_1     =   '#2ca02c', # tab:green
    mmacri_0        =   '#ffd700', # yellow
    afernandez_0    =   '#17becf', # tab:cyan
    jmilei_0        =   '#9467bd', # tab:purple
)

RGB_por_presidencia = {
    k: Color.from_hex(v).as_rgb() 
    for k, v in colores_por_presidencia.items()
}

continuous_colorizer_por_presidencia = {
    k: ContinuousColorizer.from_colors(white, v)
    for k, v in RGB_por_presidencia.items()
}

for k, v in continuous_colorizer_por_presidencia.items():
    display(v.get_palette(range(200)))
```

```python
colores_por_pregunta = dict(
    q0 = '#D00000',
    q1 = '#FFBA08',
    q2 = '#CBFF8C',
    q3 = '#8FE388',
    q4 = '#1B998B',
    q5 = '#3185FC',
    q6 = '#5D2E8C',
    q7 = '#46237A',
    q8 = '#FF7B9C',
    q9 = '#FF9B85',
)

RGB_por_pregunta = {
    k: Color.from_hex(v).as_rgb() 
    for k, v in colores_por_pregunta.items()
}

continuous_colorizer_por_pregunta = {
    k: ContinuousColorizer.from_colors(white, v)
    for k, v in RGB_por_pregunta.items()
}

for k, v in continuous_colorizer_por_pregunta.items():
    display(v.get_palette(range(200)))
```

```python
df = pl.read_csv('../data/sum_of_relevances_per_president.csv')
df
```

```python
global_colorizer = ContinuousColorizer.from_colors(red, green)
display(global_colorizer.get_palette(range(200)))
global_colorizer = global_colorizer.get_discrete_palette(df['value'].to_list())
```

```python
df = df.with_columns([
    pl.col('value')
        .map_elements(global_colorizer, return_dtype=pl.Object)
        .map_elements(lambda x: x.as_hex(), return_dtype=pl.String)
        .alias('global_color')
])
```

```python
presidentes = df['presidente'].unique(maintain_order=True).to_list()


column = [None] * len(df)
for presidente in presidentes:
    slice = df.with_row_index().filter(df['presidente'] == presidente)
    colorizer = continuous_colorizer_por_presidencia[presidente]
    f = colorizer.get_discrete_palette(slice['value'].to_list())
    for i in slice['index']:
        valor = df['value'][i]
        column[i] = f(valor).as_hex()

df = df.with_columns(
    color_norm_presidente = pl.Series(column)
)
```

```python
preguntas = df['qid'].unique(maintain_order=True).to_list()

column = [None] * len(df)
for pregunta in preguntas:
    slice = df.with_row_index().filter(df['qid'] == pregunta)
    colorizer = continuous_colorizer_por_pregunta[pregunta]
    f = colorizer.get_discrete_palette(slice['value'].to_list())
    for i in slice['index']:
        valor = df['value'][i]
        column[i] = f(valor).as_hex()
        
df = df.with_columns(
    color_norm_pregunta = pl.Series(column)
)
```

```python
df = df.to_pandas() if isinstance(df, pl.DataFrame) else df

precedence_mapping = {
    'fdelarua_0': 0,
    'eduhalde_0': 1,
    'nkirchner_0': 2,
    'ckirchner_1': 3,
    'mmacri_0': 4,
    'afernandez_0': 5,
    'jmilei_0': 6
}

presidents_order = sorted(precedence_mapping, key=precedence_mapping.get)

df['presidente'] = pd.Categorical(df['presidente'], categories=presidents_order, ordered=True)

def hex_to_rgb(hex_color):
    """Convert hex color to RGB tuple."""
    hex_color = hex_color.lstrip('#')
    return tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))

def get_luminance(rgb):
    """Calculate the luminance of an RGB color."""
    r, g, b = rgb
    return 0.299 * r + 0.587 * g + 0.114 * b

valor_pivot = df.pivot(index='presidente', columns='qid', values='value')
color_pivot = df.pivot(index='presidente', columns='qid', values='color_norm_presidente')

presidents = valor_pivot.index.tolist()
questions = valor_pivot.columns.tolist()

fig, ax = plt.subplots(figsize=(10, 8))

for i, president in enumerate(presidents):
    for j, question in enumerate(questions):
        value = valor_pivot.loc[president, question]
        color_hex = color_pivot.loc[president, question]
        if pd.isna(value) or pd.isna(color_hex):
            continue

        x = j
        y = i

        rect = plt.Rectangle((x, y), 1, 1, facecolor=color_hex)
        ax.add_patch(rect)

        # Calculate luminance to decide text color
        rgb = hex_to_rgb(color_hex)
        luminance = get_luminance(rgb)
        text_color = 'black' if luminance > 186 else 'white'
        ax.text(x + 0.5, y + 0.5, f'{value:.2f}', ha='center', va='center', color=text_color)

plt.suptitle('Suma de los puntajes de relevancia por chunk.\nColores según normalización por fila (presidente).')

ax.set_xlim(0, len(questions))
ax.set_ylim(0, len(presidents))
ax.set_xticks(np.arange(len(questions)) + 0.5)
ax.set_yticks(np.arange(len(presidents)) + 0.5)
ax.set_xticklabels(questions)
ax.set_yticklabels(presidents)
plt.xticks(rotation=0)
ax.invert_yaxis()
ax.set_aspect('equal')

plt.show()
```
